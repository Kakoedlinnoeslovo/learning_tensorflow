```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x_val = np.linspace(-1, 1, 101)
y_val = 2 * x_val + np.random.randn(*x_val.shape) * 0.33
```

```python
plt.plot(x_val, y_val)
```

```python
x = tf.placeholder("float")
y = tf.placeholder("float")
w = tf.Variable(0.0, name = "weights")

y_out = tf.multiply(x, w)

cost = tf.losses.mean_squared_error(y, y_out)

train_step = tf.train.GradientDescentOptimizer(learning_rate=0.01).\
                                                    minimize(cost)

with tf.Session() as sess:
    
    #you need to initialize variables (in this case just variable w)
    tf.global_variables_initializer().run()
    
    for i in range(100):
        for x_, y_ in zip(x_val, y_val):
            sess.run(train_step, feed_dict={x: x_, y: y_})
    print(sess.run(w)) # It should be something around 2
        
```
